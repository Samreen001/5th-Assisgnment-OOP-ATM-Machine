```python
# üè¶ ATM Machine Simulation using OOP

class ATM:
    def __init__(self):
        self.balance = 5000
        self.pin = 1234

    def check_pin(self, input_pin):
        if input_pin == self.pin:
            return True
        else:
            print("\nüîê Incorrect PIN. Please try again.")
            return False

    def check_balance(self):
        print(f"\nüí∞ Your current balance is: Rs. {self.balance}")

    def deposit(self, input_pin, amount):
        if not self.check_pin(input_pin):
            return
        if amount <= 0:
            print("\n‚ùå Deposit amount must be greater than 0.")
        else:
            self.balance += amount
            print(f"\n‚úÖ Successfully deposited Rs. {amount}.")
            self.check_balance()

    def withdraw(self, input_pin, amount):
        if not self.check_pin(input_pin):
            return
        if amount <= 0:
            print("\n‚ùå Withdrawal amount must be greater than 0.")
        elif amount > self.balance:
            print("\n‚ùå Insufficient balance.")
        else:
            self.balance -= amount
            print(f"\n‚úÖ Successfully withdrawn Rs. {amount}.")
            self.check_balance()

    def exit(self):
        print("\nüëã Thank you for using the ATM. Goodbye!")
        exit()

# üöÄ Start the ATM
atm = ATM()

# üìã Menu-driven interface
while True:
    print("\n" + "="*35)
    print("          üèß ATM MAIN MENU")
    print("="*35)
    print("1Ô∏è‚É£  Check Balance")
    print("2Ô∏è‚É£  Deposit Money")
    print("3Ô∏è‚É£  Withdraw Money")
    print("4Ô∏è‚É£  Exit")
    print("="*35)

    choice = input("üëâ Enter your choice (1-4): ")

    if choice == '1':
        try:
            pin = int(input("üîë Enter your PIN: "))
            if atm.check_pin(pin):
                atm.check_balance()
        except ValueError:
            print("‚ùó Please enter a valid PIN (numbers only).")

    elif choice == '2':
        try:
            pin = int(input("üîë Enter your PIN: "))
            amount = float(input("üíµ Enter amount to deposit: "))
            atm.deposit(pin, amount)
        except ValueError:
            print("‚ùó Please enter a valid amount.")

    elif choice == '3':
        try:
            pin = int(input("üîë Enter your PIN: "))
            amount = float(input("üí∏ Enter amount to withdraw: "))
            atm.withdraw(pin, amount)
        except ValueError:
            print("‚ùó Please enter a valid amount.")

    elif choice == '4':
        atm.exit()

    else:
        print("‚ùó Invalid choice. Please select option 1 to 4.")
```
